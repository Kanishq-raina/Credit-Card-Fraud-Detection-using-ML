```python
#@title Filling the null spaces in dataframe with previous value
credit_card_data2=credit_card_data.fillna(method='pad')



```

**WE CAN SEE THAT DATA DISTRIBUTION IS HIGHLY UNBALANCED**

```python
#@title Seperating the data for analysis
legit=credit_card_data2[credit_card_data2.Class==0]
fraud=credit_card_data2[credit_card_data2.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

**BUILD A SAMPLE DATASET CONTAINING SIMILIAR DISTRIBUTION OF NORMAL TRANSACTOINS AND FRAUDULENT TRANSACTIONS
**

```python
legit_sample=legit.sample(n=52)
```

```python
#@title Concatinating two Dataframes
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
#@title New dataset
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#@title Splitting the data into features and targets
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
print(X)
```

```python
#@title SPLITTING DATA INTO TRAINING DATA AND TESTING DATA
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#@title Model training & Logistic Regression
model=LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
#@title Training the Logistic Regression Model with Training Data
model.fit(X_train,Y_train)
```

**Model Evaluation**

```python
#@title Accuracy Score
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data: ',training_data_accuracy)
```

```python
#@title Accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print('Accuracy score on test data: ',test_data_accuracy)
```
